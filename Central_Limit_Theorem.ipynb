```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean
from scipy import stats
from numpy.random import seed
```

Complete these steps for two of the following distributions:

Normal
Binomial
Gamma
Poisson

Create a line chart that plots the sample mean as more and more trials are conducted, up to 1,000 trials.

Randomly sample the mean 1,000 times, and plot the resulting frequency of sample means as a histogram.

```python
results = []
for num_trials in range(1,1000):
  trials = np.random.normal(5, 0.5, num_trials)
  mean_trials = trials.mean()
  results.append(mean_trials)
df = pd.DataFrame({ 'trials' : results})

df.plot()
```

```python
results = []
for num_trials in range(1,1000):
  trials = np.random.binomial(30, 0.7, num_trials)
  mean_trials = trials.mean()
  results.append(mean_trials)
df = pd.DataFrame({ 'trials' : results})

df.plot()
```

```python
results = []
for num_trials in range(1,1000):
  trials = np.random.gamma(2, 0.2, num_trials)
  mean_trials = trials.mean()
  results.append(mean_trials)
df = pd.DataFrame({ 'trials' : results})

df.plot()
```

```python
results = []
for num_trials in range(1,1000):
  trials = np.random.poisson(6, num_trials)
  mean_trials = trials.mean()
  results.append(mean_trials)
df = pd.DataFrame({ 'trials' : results})

df.plot()
```
